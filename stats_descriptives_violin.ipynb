```python
# import libraries
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

```python
## create the data

n = 1000
thresh = 5 # threshold for cropping data
data = np.exp(np.random.randn(n))
data[data > thresh] = thresh + np.random.randn(sum(data>thresh)) * .1

# show histogram
plt.hist(data, 30)
plt.title('Histogram')
plt.show()

# show violin plot
plt.violinplot(data)
plt.title('violin')
plt.show()
```

```python
# aonther option: swarm plot
import seaborn as sns
sns.swarmplot(data, orient='v')
```

```python
## create the data

n = 1000
thresh = 5 # threshold for cropping data
data1 = np.exp(np.random.randn(n))
data1[data1 > thresh] = thresh + np.random.randn(sum(data1>thresh)) * .1

data2 = np.random.randn(n)
data2[data2 > thresh] = thresh + np.random.randn(sum(data2>thresh)) * .1

# show histogram
plt.hist(data1)
plt.title('Histogram')
plt.show()

plt.hist(data2)
plt.title('data2 histogram')
plt.show()

# show violin plot
plt.violinplot([data1, data2])
plt.title('violin')
plt.show()
```
